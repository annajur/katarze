```python
# 
a = np.array([[BB, BV, BR], [BV, VV, VR], [BR, VR, RR]])

bx = np.array([XB, XV, XR])
solx = np.linalg.solve(a, bx)

by = np.array([YB, YV, YR])
soly = np.linalg.solve(a, by)

bz = np.array([ZB, ZV, ZR])
solz = np.linalg.solve(a, bz)

print(a)

print(solx)
print(soly)
print(solz)
```

### Kyzena matice:

```python
matr = np.array([solx, soly, solz])
print(matr)
```

```python
img_b = apf.getdata("A2F5.fits")
img_v = apf.getdata("A2F4.fits")
img_r = apf.getdata("A2F1.fits")
```

```python
#slozeni v jeden, otoceni (a zakomentovane preklopeni)
img_bvr = np.dstack((img_b, img_v, img_r))
img_bvr = np.rot90(img_bvr)
#img_bvr = img_bvr[:,::-1,:]
```

```python
plt.imshow(img_bvr)
```

```python
#funkce zajistujici transformaci pomoci sestavene matice; prvni arg obrazek, druhy arg matice
def trans(img, m):
    tmp = np.copy(img.astype(float))
    for i in range(tmp.shape[0]):
        for j in range(tmp.shape[1]):
            tmp[i,j] = m.dot(tmp[i,j])
    return tmp
```

```python
normalize = np.vectorize(lambda x, x0, x1: (x - x0)/ (x1 - x0))
```

```python
img_1 = np.copy(img_bvr.astype(float))

img_1[:,:,0] = normalize(img_1[:,:,0],50.0, 340.0) #maximalni hodnoty pro normalizaci jsou stanoveny na zaklade kvalifikovaneho odhadu (proste nejak zhruba)
img_1[:,:,1] = normalize(img_1[:,:,1],0.0, 1900.0)
img_1[:,:,2] = normalize(img_1[:,:,2],0.0, 2500.0)

img_2 = trans(img_1, matr)
plt.figure(figsize=(8,8))
plt.imshow(img_2)
```

```python
m_xyz_srgb = np.array([
        [3.2404542, -1.5371385, -0.4985314],
        [-0.9692660,  1.8760108,  0.0415560],
        [0.0556434, -0.2040259,  1.0572252]
    ])
```

```python
img_1_n = trans(img_2,m_xyz_srgb)
img_1_n[:,:,0] = normalize(img_1_n[:,:,0],0, .45*img_1_n[:,:,0].max()) #max hodnot bylo dosazeno stejne jako v predchozim pripade
img_1_n[:,:,1] = normalize(img_1_n[:,:,1],0, .9*img_1_n[:,:,0].max())
img_1_n[:,:,2] = normalize(img_1_n[:,:,2],0, 0.9*img_1_n[:,:,0].max())
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img_1_n)
```

### Po linearnim prevodu z XYZ do sRGB je potreba jeste pouzit $\gamma$-funkci, diky niz jsou lepe videt detaily... a udela se sedy zakal. :D Ale aspon je videt cerveny vnitrek Matejova batohu -- a ze cervena je skutecne cervena.

```python
img_1_ng = np.copy(img_1_n.astype(float))
img_1_ng[:,:,0] = gama(img_1_ng[:,:,0])
img_1_ng[:,:,1] = gama(img_1_ng[:,:,1])
img_1_ng[:,:,2] = gama(img_1_ng[:,:,2])
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img_1_ng)
```

#### S tou tastickou, co by mela byt spis cernohneda, se neda nic moc delat, rekla bych, ze je to tim, ze R filtr ma rozsah dal do IR, takze je CCDckem lepe videt.

```python
def gama(x):
    tmp = np.zeros_like(x, dtype = float)
    mask = x>3.1308e-3
    tmp[mask] = x[mask]**(1/2.4)*1.055-0.055
    tmp[mask == False] = x[mask == False]*12.92
    return tmp
```

```python
img_1_ngn = np.copy(img_1_n.astype(float))
img_1_ngn[:,:,0] = normalize(img_1_ng[:,:,0],-0.0, 1/1.5)#img_1[:,:,2].max())
img_1_ngn[:,:,1] = normalize(img_1_ng[:,:,1],-0.0, 1/1.5)#img_1[:,:,2].max())
img_1_ngn[:,:,2] = normalize(img_1_ng[:,:,2],-0.0, 1/1.5)#img_1[:,:,2].max())
```
